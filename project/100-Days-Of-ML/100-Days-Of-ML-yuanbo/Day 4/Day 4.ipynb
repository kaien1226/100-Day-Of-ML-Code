# Step 1: Data Pre-Processing

## importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## importing the dataset

```python
dataset =  pd.read_csv('/home/zhengyuanbo/project/100-Days-Of-ML/100-Days-Of-ML-yuanbo/datasets/Social_Network_Ads.csv')
print(dataset)
```

```python
X = dataset.iloc[:, 2:4].values
print(X)
Y = dataset.iloc[:, 4].values
print(Y)
```

## splitting the dataset into the train set and teat set

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25,random_state = 0)
```

# feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_test)
```

# Step2 : Logistic Regression Model

## Fitting logistic regression to the train set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
```

# Step 3: Predection

## predicting the test set results

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

# Step 4: Evaluatig the predection

## Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm  = confusion_matrix(Y_test, y_pred)
print(cm)
```

## Visualization

```python
def plot_decision_boundary(model,axis):    
    x0,x1 = np.meshgrid(         
        np.linspace(axis[0],axis[1],int((axis[1]-axis[0])*100)),         
        np.linspace(axis[2],axis[3],int((axis[3]-axis[2])*100))     
    )     
    X_new = np.c_[x0.ravel(),x1.ravel()]          
    y_predict = model.predict(X_new)     
    zz = y_predict.reshape(x0.shape)        
    from matplotlib.colors import ListedColormap     
    custom_cmap = ListedColormap(['#EF9A9A','#9A9AFF'])          
    plt.contourf(x0,x1,zz,linewidth=5,cmap=custom_cmap)
```

```python
x_train_coordinate = X_train[:, 0]
y_train_coordinate = X_train[:, 1]

x_test_coordinate = X_test[:, 0]
y_test_coordinate = X_test[:, 1]

plt.figure(figsize = (15,5))
plt.subplot(121)
plot_decision_boundary(classifier,axis=[-2.5,3.5,-2,3])
plt.scatter(x_train_coordinate[Y_train == 0], y_train_coordinate[Y_train == 0], color = 'b', marker = 'o', label = 'Y=0')
plt.scatter(x_train_coordinate[Y_train == 1], y_train_coordinate[Y_train == 1], color = 'r',marker = '+', label = 'Y=1')
plt.xlabel('Age')
plt.ylabel('salay')
plt.legend(loc=1)
plt.title('Train set')

#plt.figure(2)
plt.subplot(122)
plot_decision_boundary(classifier,axis=[-2.5,3.5,-2,3])
plt.scatter(x_test_coordinate[Y_test == 0], y_test_coordinate[Y_test == 0], color = 'b', marker = 'o', label = 'Y=0')
plt.scatter(x_test_coordinate[Y_test == 1], y_test_coordinate[Y_test == 1], color = 'r',marker = '+', label = 'Y=1')
plt.xlabel('Age')
plt.ylabel('salay')
plt.legend(loc=1)
plt.title('Test set')

plt.show()
```
